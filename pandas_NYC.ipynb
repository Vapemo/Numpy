```python
import pyarrow.parquet as pq
import pandas as pd


#como no pude descargar el dataset seleccione las columnaas necesarias
columnas_necesarias = [
    "pickup_datetime",
    "trip_miles",
    "base_passenger_fare",
    "tips",
    "tolls",
    "driver_pay"
]

fhvhv_taxi = pq.read_table("/content/fhvhv_tripdata_2025-01(2).parquet", columns=columnas_necesarias)
fhvhv_df = fhvhv_taxi.to_pandas()

fhvhv_df["pickup_datetime"] = pd.to_datetime(fhvhv_df["pickup_datetime"])
fhvhv_df["pickup_hour"] = fhvhv_df["pickup_datetime"].dt.hour
fhvhv_df["pickup_day"] = fhvhv_df["pickup_datetime"].dt.day

#renombre de las columnas
fhvhv_df = fhvhv_df.rename(columns={
    "trip_miles": "trip_distance",
    "base_passenger_fare": "total_amount",
    "tips": "tip_amount",
    "tolls": "tolls_amount"
})

print(fhvhv_df.head())
```

#**Promedio de los ingresos brutos  de casa taxi y de todos los taxis**

```python
#promedio de los taxis amarillos
valid_taxi_yelllow = taxi_yellow[taxi_yellow["total_amount"] > 0]
promedio_yellow = float(taxi_yellow["total_amount"].mean())

print( f"Promedio ingresos brutos (yellow_taxis): {promedio_yellow} ")

#promedio de los taxis verdes
valid_taxi_green= taxi_green[taxi_green["total_amount"] > 0]
promedio_green = float(taxi_green["total_amount"].mean())

print( f"Promedio ingresos brutos (green_taxis):: {promedio_green}")
#promedio de fhvhv
valid_fhv = fhvhv_df[fhvhv_df["total_amount"] > 0]
promedio_fhv = float(valid_fhv["total_amount"].mean())
print(f"Promedio de los ingresos brutos de los FHV es: {promedio_fhv:.2f}")
```

```python
#Promedio de todos los taxis

promedio_total=(promedio_yellow + promedio_green+ promedio_fhv)/3
print(f"El promedio total es:{promedio_total}")
```






#**Tipo  de taxis recorrieron la mayor cantidad de distancia en Enero**

```python

# distancia total recorrida
yellow_distance = taxi_yellow["trip_distance"].sum()
green_distance = taxi_green["trip_distance"].sum()
hvfh_distance = fhvhv_df["trip_distance"].sum()

# comparación de  distancias
distances = {
    "Yellow Taxi": yellow_distance,
    "Green Taxi": green_distance,
    "HVFH_Taxi": hvfh_distance

}

taxi_mas_distancia = max(distances, key=distances.get)

print(f"El tipo de taxi que recorrió la mayor distancia en enero es: {taxi_mas_distancia}")
print(f"Distancias Totales:")
for taxi_type, distance in distances.items():
    print(f"- {taxi_type}: {distance} millas")
```

```python
#promedio de propinas
yellow_tips =  taxi_yellow["tip_amount"].mean()
green_tips = taxi_green["tip_amount"].mean()
hvfh_tips = fhvhv_df["tip_amount"].mean()

#promedios
tips_avg = {
    "Yellow Taxi": yellow_tips,
    "Green Taxi": green_tips,
    "HVFH Taxi": hvfh_tips
}

#  taxi con mayores propinas en promedio
taxi_mas_propinas = max(tips_avg, key=tips_avg.get)

print(f"Taxi con más propinas en promedio: {taxi_mas_propinas}")
print("Propinas promedio:")
print(tips_avg)
```

#***Gráfico de barras reportando el importe del peaje total de cada tipo de Taxi***

```python
#Gráfico de barras reportando el importe del peaje total de cada tipo de Taxi
import matplotlib.pyplot as plt

peaje = {
    "Yellow Taxi": taxi_yellow["tolls_amount"].sum(),
    "Green Taxi": taxi_green["tolls_amount"].sum(),
    "HVFH Taxi": fhvhv_df["tolls_amount"].sum()
}

plt.figure(figsize=(10, 6))
bars = plt.bar(peaje.keys(), peaje.values(), color=["#FFD700", "#32CD32", "#4682B4"])

# Etiquetas y título
plt.title("Total de Peajes por Tipo de Taxi - Enero 2025", fontsize=14, pad=20, fontweight='bold')
plt.xlabel("Tipo de Taxi", fontsize=12)
plt.ylabel("Importe- USD", fontsize=12)

#etiquetas
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'${height:,.0f}', ha='center', va='bottom')

# ajuste de y
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x:,.0f}'))

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#***Mapa de calor que describa el comportamiento de a qué hora se tomaron todos los taxis durante los días de Enero***

```python

import matplotlib.pyplot as plt
import seaborn as sns

yellow_pickups = pd.to_datetime(taxi_yellow["tpep_pickup_datetime"])
green_pickups = pd.to_datetime(taxi_green["lpep_pickup_datetime"])
fhvhv_pickups = pd.to_datetime(fhvhv_df["pickup_datetime"])

all_pickups = pd.concat([yellow_pickups, green_pickups, fhvhv_pickups], ignore_index=True)

heatmap = pd.DataFrame({
    "day": all_pickups.dt.day,
    "hour": all_pickups.dt.hour
})

pivot_table = heatmap.pivot_table(
    index="day",
    columns="hour",
    aggfunc="size",
    fill_value=0
)

plt.figure(figsize=(14, 8))
sns.heatmap(
    pivot_table,
    cmap="Blues",  #YlGnBu
    annot=True,
    fmt="d",
    linewidths=0.5,
    cbar_kws={"label": "Número de Viajes"}
)

plt.title("Comportamiento de Horario de Viajes en Taxis - Enero 2025", fontsize=18, pad=20, fontweight='bold')
plt.xlabel("Hora ", fontsize=12)
plt.ylabel("Día de Enero", fontsize=12)

plt.tight_layout()
plt.show()

```

# ***Mapa de calor para cada tipo de taxi que reporte los días de Enero en los que más se uso su servicio***

```python
import matplotlib.pyplot as plt
import seaborn as sns

yellow_pickups = pd.to_datetime(taxi_yellow["tpep_pickup_datetime"])
yellow_heatmap = pd.DataFrame({
    "day": yellow_pickups.dt.day,
    "hour": yellow_pickups.dt.hour
})

yellow_pivot_table = yellow_heatmap.pivot_table(
    index="day", columns="hour", aggfunc="size", fill_value=0
)

plt.figure(figsize=(14, 8))
sns.heatmap(
    yellow_pivot_table,
    cmap="YlOrBr",
    annot=False,
    fmt="d",
    linewidths=0.5,
    cbar_kws={"label": "Número de Viajes"}
)

plt.title("Yellow Taxis: Uso por Día y Hora- Enero 2025", fontsize=18, pad=20,fontweight='bold')
plt.xlabel("Hora", fontsize=12)
plt.ylabel("Día de Enero", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
green_pickups = pd.to_datetime(taxi_green["lpep_pickup_datetime"])
green_heatmap = pd.DataFrame({
    "day": green_pickups.dt.day,
    "hour": green_pickups.dt.hour
})

green_pivot_table = green_heatmap.pivot_table(
    index="day", columns="hour", aggfunc="size", fill_value=0
)

plt.figure(figsize=(14, 8))
sns.heatmap(
    green_pivot_table,
    cmap="Greens",
    annot=False,
    fmt="d",
    linewidths=0.5,
    cbar_kws={"label": "Número de Viajes"}
)


plt.title("Green Taxis: Uso por Día y Hora- Enero 2025", fontsize=18, pad=20, fontweight='bold')
plt.xlabel("Hora ", fontsize=12)
plt.ylabel("Día de Enero", fontsize=12)
plt.tight_layout()
plt.show()
```
