#**Promedio de ingresos brutos de cada tipo de taxi y de todos los taxis**

```python

promedio_yellow= yellow_taxi.select(pl.col("total_amount").mean()).to_numpy()
print(f"Promedio yellow taxis:{promedio_yellow}")
promedio_green=green_taxi.select(pl.col("total_amount")).mean().to_numpy()
print(f"Promedio greeen taxis:{promedio_green}")

fhvhv = fhvhv.with_columns(
    (
        pl.col("base_passenger_fare") +
        pl.col("tolls") +
        pl.col("sales_tax") +
        pl.col("congestion_surcharge") +
        pl.col("airport_fee") +
        pl.col("tips")
    ).alias("total_amount")
)


fhvhv_promedio = fhvhv.select(pl.col("total_amount").mean()).to_numpy()[0][0]

print(f"Promedio FHVhV taxis: {fhvhv_promedio:.2f}")
promedio_general=(promedio_yellow + promedio_green +  fhvhv_promedio)/3
print(f"Promedio general de los ingresos brutos {promedio_general}")
```

#**Tipo de taxis recorrieron la mayor cantidad de distancia en Enero**

```python

distancia_yellow = yellow_taxi["trip_distance"].sum()
distancia_green = green_taxi["trip_distance"].sum()
distancia_fhvhv = fhvhv["trip_miles"].sum()


df_distancias = pl.DataFrame({
    "Taxi Type": ["Yellow Taxi", "Green Taxi", "FHVs"],
    "Distance (miles)": [distancia_yellow, distancia_green, distancia_fhvhv]
})

taxi_mas_distancia = df_distancias["Taxi Type"][df_distancias["Distance (miles)"].arg_max()]
max_distance = df_distancias["Distance (miles)"].max()


print("DISTANCIAS TOTALES RECORRIDAS EN ENERO")

print(df_distancias)

print(f"\n Taxi con mayor distancia recorrida es {taxi_mas_distancia} ({max_distance:,.2f} millas)")
```

```python
tip_yellow = yellow_taxi["tip_amount"].mean()
tip_green = green_taxi["tip_amount"].mean()
tip_fhvhv = fhvhv["tips"].mean()

df_prom_tips = pl.DataFrame({
    "Taxi Type": ["Yellow Taxi", "Green Taxi", "FHVs"],
    "Average Tip": [tip_yellow, tip_green, tip_fhvhv]
})

taxi_mas_tip = df_prom_tips["Taxi Type"][df_prom_tips["Average Tip"].arg_max()]
max_tip = df_prom_tips["Average Tip"].max()


total_tip_yellow = yellow_taxi["tip_amount"].sum()
total_tip_green = green_taxi["tip_amount"].sum()
total_tip_fhvhv = fhvhv["tips"].sum()


df_total_tips = pl.DataFrame({
    "Taxi Type": ["Yellow Taxi", "Green Taxi", "FHVs"],
    "Total Tips": [total_tip_yellow, total_tip_green, total_tip_fhvhv]
})


print("\n Promedio de propinas por tipo de taxi:")
print(df_prom_tips)

print("\n Propinas totales por tipo de taxi:")
print(df_total_tips)

print(f" Taxi con más propinas en promedio: {taxi_mas_tip} (${max_tip:.2f})")
print(f" Yellow Taxis - Total: ${total_tip_yellow:,.2f}")
print(f" Green Taxis - Total: ${total_tip_green:,.2f}")
print(f" FHVs - Total: ${total_tip_fhvhv:,.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#**Gráfico de barras reportando el importe del peaje total de cada tipo de Taxi**

```python


tolls_data = [
    ("Yellow Taxi", yellow_taxi["tolls_amount"].sum()),
    ("Green Taxi", green_taxi["tolls_amount"].sum()),
    ("FHVs", fhvhv["tolls"].sum())
]

df_tolls = pl.DataFrame({
    "Taxi Type": [x[0] for x in tolls_data],
    "Total Tolls": [x[1] for x in tolls_data]
})


plt.figure(figsize=(10, 6))
bars = plt.bar(df_tolls["Taxi Type"], df_tolls["Total Tolls"],
               color=['#FFD700', '#008000', '#4682B4'])

plt.title("Total de Peajes por Tipo de Taxi - Enero 2025", fontsize=14, pad=20, fontweight='bold')
plt.xlabel("Tipo de Taxi", labelpad=10)
plt.ylabel("Importe-USD", labelpad=10)
plt.grid(axis='y', linestyle='--', alpha=0.3)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'${height:,.0f}',
             ha='center', va='bottom')

plt.gca().yaxis.set_major_formatter('${x:,.0f}')
plt.tight_layout()
plt.show()
```

#***Mapa de calor que describa el comportamiento de a qué hora se tomaron todos los taxis durante los días de Enero***

```python


taxis = {
    'Yellow Taxis': (yellow_taxi, 'tpep_pickup_datetime'),
    'Green Taxis': (green_taxi, 'lpep_pickup_datetime'),
    'FHVHV': (fhvhv, 'pickup_datetime')
}

all_pickups = (
    yellow_taxi['tpep_pickup_datetime']
    .append(green_taxi['lpep_pickup_datetime'])
    .append(fhvhv['pickup_datetime'])
)

heatmap= all_pickups.to_frame().with_columns([
    all_pickups.dt.hour().alias("hour"),
    all_pickups.dt.day().alias("day")
])

pivot_table = heatmap.group_by(["day", "hour"]).agg(pl.len()).sort(
    "day", "hour"
).rename(
    {'len': 'counts'}).pivot(
    index="day",
    on="hour",
    values="counts",
).fill_null(0)

sns.set_theme(style="whitegrid")
plt.figure(figsize=(16, 16))

ax = sns.heatmap(
    pivot_table.drop('day').to_pandas(),
    cmap='Blues',
    linewidths=0.1,
    linecolor='white',
    annot=True,
    fmt=".0f",
    cbar_kws={'label': 'Número de Viajes', 'shrink': 0.8}
)

plt.title('Comportamiento de Horario de Viajes en Taxis - Enero 2025',
         fontsize=20, pad=20, fontweight='bold')
plt.xlabel('Hora ', fontsize=12)
plt.ylabel('Día de Enero', fontsize=12)


plt.tight_layout()
plt.show()
```

# ***Mapa de calor para cada tipo de taxi que reporte los días de Enero en los que más se uso su servicio***

```python


yellow_heatmap = yellow_taxi['tpep_pickup_datetime'].to_frame().with_columns([
    yellow_taxi['tpep_pickup_datetime'].dt.hour().alias("hour"),
    yellow_taxi['tpep_pickup_datetime'].dt.day().alias("day")
])

yellow_pivot = yellow_heatmap.to_pandas().pivot_table(
    index="day", columns="hour", aggfunc="size", fill_value=0
)

plt.figure(figsize=(12, 6))
sns.heatmap(
    yellow_pivot,
    cmap="YlOrBr",
    annot=False,
    fmt=".0f",
    linewidths=0.5,
    cbar_kws={'label': 'Número de Viajes'}
)
plt.title("Yellow Taxis: Uso por Día y Hora-Enero 2025", fontsize=16, pad=15,fontweight='bold')
plt.xlabel("Hora ", fontsize=12)
plt.ylabel("Día de Enero", fontsize=12)
plt.tight_layout()
plt.show()

```

```python

green_heatmap = green_taxi['lpep_pickup_datetime'].to_frame().with_columns([
    green_taxi['lpep_pickup_datetime'].dt.hour().alias("hour"),
    green_taxi['lpep_pickup_datetime'].dt.day().alias("day")
])

green_pivot = green_heatmap.to_pandas().pivot_table(
    index="day", columns="hour", aggfunc="size", fill_value=0
)

plt.figure(figsize=(12, 6))
sns.heatmap(
    green_pivot,
    cmap="Greens",
    annot=False,
    fmt=".0f",
    linewidths=0.5,
    cbar_kws={'label': 'Número de Viajes'}
)
plt.title("Green Taxis: Uso por Día y Hora-Enero 2025", fontsize=16, pad=15,fontweight='bold')
plt.xlabel("Hora ", fontsize=12)
plt.ylabel("Día de Enero", fontsize=12)
plt.tight_layout()
plt.show()
```
